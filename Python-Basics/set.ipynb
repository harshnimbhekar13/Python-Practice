```python
set1={'Apple','Banana','Grapes'}
set2={'Microsoft','Apple','Nokia'}
```

```python
set3=set1.intersection(set2) # Intersection method are support only same element
```

```python
set3
```

```python
set4=set1.symmetric_difference(set2) # Does not support same element 
```

```python
set4
```
