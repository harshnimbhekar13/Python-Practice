# What is Strings

#In python,anything that you enclose between single or double quotation markd is considered a string. A string is essentially a sequence or array of textual data. String are used when working with Unicode characters.

```python
# Example
name="harsh"
print('Hey,' +name)
```

# Accessing Characters of a String 

```python
# In Python, string is like an array of characters. we can access parts of string by using its index which starts from 0.
# Square brackets can be used to access elements of the string.
```

```python
print(name[0])
print(name[2])
```

# Looping through the string

```python
# we can loop through string using a for loop like this:
```

# String Slicing & Operations on String

Length of a string

```python
# We can find the lenght of a string using len() Function .
```

```python
# Example
fruit='Mango'
len1=len(fruit)
print('Mango is a ',len1,'letter word.')
print(fruit[0:4]) # including 0 but not 4
print(fruit[-3:-1])
```

# String as an array

A string is essentially a sequence of characters also called an array. Thus we can access the element of this array.

```python
# Example
nm='Harsh'
print('First character : ',nm[0])
print('Last character : ',nm[-1])
```
