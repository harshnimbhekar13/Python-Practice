# String Slicing & Operations on String

Length of a string

```python
# We can find the lenght of a string using len() Function .
```

```python
# Example
fruit='Mango'
len1=len(fruit)
print('Mango is a ',len1,'letter word.')
print(fruit[0:4]) # including 0 but not 4
print(fruit[-3:-1])
```

# Split():

The split () method split the given string at the specified instance and return the separated strings as list item.

```python
str1= "Golden Spoon"
print(str1.split(" ")) # splits the string at the whitespace "  ".
```

# Capitalize()

The capitalize() method turns only the first characterof the string to uppercase and the rest other characters of the string are turned to lowercase. The string has no effect if the first character is already uppercase.

```python
str2="data anaLyst"
print(str2.capitalize())
```

# Center Method

The center() Method aligns the string to the center as per the parameters given by the user. 

```python
str1 = "Welcome to the Console!!"
print(str1.center(50))
```

# Count()

The count() method returns the number of times the given value has occurred within the given string.

```python
string = "harsh"
countstring = string.count("h")
print(countstring)
```

# Endswith():

The endswith() method checks if the string ends with a given value. If yes then return True, else return False.x

```python
str1 = "Welcome to the Console!!"
print(str1.endwith("!!!"))
```
