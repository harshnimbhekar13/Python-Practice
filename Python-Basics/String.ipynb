# String Slicing & Operations on String

Length of a string

```python
# We can find the lenght of a string using len() Function .
```

```python
# Example
fruit='Mango'
len1=len(fruit)
print('Mango is a ',len1,'letter word.')
print(fruit[0:4]) # including 0 but not 4
print(fruit[-3:-1])
```

# Split():

The split () method split the given string at the specified instance and return the separated strings as list item.

```python
str1= "Golden Spoon"
print(str1.split(" ")) # splits the string at the whitespace "  ".
```

# Capitalize()

The capitalize() method turns only the first characterof the string to uppercase and the rest other characters of the string are turned to lowercase. The string has no effect if the first character is already uppercase.

```python
str2="data anaLyst"
print(str2.capitalize())
```

# Center Method

The center() Method aligns the string to the center as per the parameters given by the user. 

```python
str1 = "Welcome to the Console!!"
print(str1.center(50))
```

# Count()

The count() method returns the number of times the given value has occurred within the given string.

```python
string = "harsh"
countstring = string.count("h")
print(countstring)
```

# Find():

The find () method searches for the first occurrence of the given value and returns the index where it is present. If given value is absent from the string then return -1.

```python
str1="He's name is Don. He is an honest man."
print(str1.find("is"))
```

# Index()

The index() method searches for the first occurrence of the given value and 
return the index where it is present. If given value is absent from the string 
then raise an exception.

```python
str1="He's name is Don. He is an honest man."
print(str1.index("Daniel"))
```

# isalnum():

The isalnum() method returns True only if the entire string only consists
of A-Z,a-z,0-9. If anyn other characters or punctuactions are present, then
it returns False.

```python
str1 = "Welcome to the Console!!"
print(str1.isalnum())
```

# isalpha():

The isalpha() method returns True only if the entire string only consists
of A-Z,a-z. If anyn other characters or punctuactions or numbers(0-9),and
present,then it returns False.

```python
str1 = "Welcome"
print(str1.isalpha())
```

# islower():

The islower() method returns True if all the characters in the string are lower case, else it return False.

```python
str1 = "Hello world"
print(str1.islower())
```

# isprintable():

The isprintable() method returns True if all values within the gives string 
are printable, if not,then return False.

```python
str1 = " Wish you a very Happy Birthday"
print(str1.isprintable())
```

# isspace():

The inspace() method returns True only if the string contains 
white speces,else returns False.

```python
str1 = "           "  # using Specebar
print(str1.isspace())
str2 = "           " # using Tab
print(str2.isspace())
```

# istitle():

The istitle() returns True only if the first letter of each word of the 
string is capitalized,else it returns False.

```python
str1 = "World Health Organization"
print(str1.istitle())
```

# isupper():

The isupper() returns True  if all the character in the string are  
upper case(),else it returns False.

```python
str1 = "World Health Organization"
print(str1.isupper())
```

# Swapcase():

The swapcase() method chnages the chracters the character casing of the string.
upper case converted to lower case and lower case to upper case.

```python
str1 = "HArSH"
print(str1.swapcase())
```
