# List Index

# Each item / element in a list has its own unique index. This index can be used to access any particular item from the list.
# The first item has index [0].second item has index [1], third item has index [2] ansd so on

```python
# Example
fruits = ['Apple','Banana','Mango','Orange']
print(fruits[0]) # Apple
print(fruits[1]) # Banana
print(fruits[2]) # Mango
print(fruits[3]) # Orange
```

# Negative Indexing 

```python
# Example
print(fruits[-1]) # Orange
print(fruits[-2]) # Mango
```

# Real life example

```python
students = ['Ritika','Harsh','Anurag','Ayan']
print('First Student : ',students[0])
print('Second Student : ',students[1])
print('Last Student : ', students[-1])
```

# Range of index 

# You can Print a range of list items by specifying where you want to start, where do you want to end and if you want to skip elements in between the range

```python
# Example
fruits = ['Apple','Banana','Mango','Orange','Grapes','Pineapple']
# Start index = 1, End Index = 5
print(fruits[1:5])

# Start index = 0, End index= 6, Skip every 2nd element 
print(fruits[0:6:2])
```

# Real-world example

```python
marks = [45,67,89,76,54,92,81]
print('Middle marks :',marks[2:5])
print('Alternate marks : ',marks[::2])
```
