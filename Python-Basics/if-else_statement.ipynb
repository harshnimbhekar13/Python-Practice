# if-else Statment

Sometimes the programmer needs to check the evaluation of creatin
expression(s),whether the expression(s) evaluate to True or False
expression evaluates to False, then the program execution follows a
different path than it would have if the expression had evaluated to true

Based on this, the conditional statements are further classified into following types:

- if
- if-else
- if-else-elif
- nested if-else-elif

# An if...else statement evaluates like this:

if the expression evalutes True:

Execute the block of code inside if statement. After execution return to the code
out of the if....else block \

if the expression evalutes False:

Execute the block of code inside else statement. After execution return to the code
out of the if....else block \

```python
a = int(input("Enter your age : "))
print("Your age is : ",a)

if(a>18):
  print("You can drive")
else:
  print("You can drive")
```

```python
applePrice = 210
budget = 200
if (applePrice <= budget):
    print("Ramu, add 1kg Apple to the cart.")
else:
    print("Ramu, do not add Apples to the card.")
```
